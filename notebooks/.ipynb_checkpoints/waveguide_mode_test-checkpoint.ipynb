```python
def make_straight_wg(shape, wg_eps, wg_thickness, wg_height):
    eps = np.ones(shape)
    i_start = eps.shape[0]//2 - wg_thickness//2
    i_end = eps.shape[0]//2 + wg_thickness//2
    k_start = eps.shape[2]//2 - wg_height//2
    k_end = eps.shape[2]//2 + wg_height//2
    eps[i_start:i_end,:,k_start:k_end] = wg_eps

    return eps
```

```python
full_shape = (64,64,64)
wg_eps = 8
wg_thickness = 20
wg_height = 20
eps, overlap_e, J = make_straight_wg(full_shape, wg_eps, wg_thickness, wg_height)
```
